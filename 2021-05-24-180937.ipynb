```python
d = {0: "black", 1: "red",2: "green", 3: "blue",4:"yellow"}

def dist(u,v):
     h = [(u[i]+v[i])%2 for i in range(len(u))]
     return sum(h)

def cube(n):
     v = {}
     G = Graph(2**n)
     vertices = Tuples([0,1],n)
     for i in range(2**n):
         for j in range(2**n):
             if dist(vertices[i],vertices[j]) == 1:
                    G.add_edge(i,j)
                    for x in range(n):
                        if vertices[i][x] == vertices[j][x]:
                            pass
                        else:
                            G.set_edge_label(i,j,x)
     for i in range(2**n):
        v[i] = (tuple(vertices[i]))
     G.relabel(v)
     return G

L = cube(2)
v_colours = {"white": [], 'black': []}
for vertex in L.vertices():
    if (sum(vertex)%2) == 0:
        v_colours["white"].append(vertex)
    else:
        v_colours["black"].append(vertex)

L.show(layout = None,edge_colors=L._color_by_label(d),vertex_colors = v_colours)
L.delete_vertex((0, 0))
L.delete_edge((0, 0),(1, 0), None)
L.delete_edge((0, 0),(0, 1), None)
L.show()
```

```python
d = {0: "black", 1: "red",2: "green", 3: "blue",4:"yellow",5:'purple'}

def add_bin(x,y):
    sum = ''
    for i in range(len(x)):
        if (int(x[i])+int(y[i])) != 2:
            sum += str(int(x[i])+int(y[i]))
        else:
            sum += "0"
    return sum

def dist(u,v):
     h = [(u[i]+v[i])%2 for i in range(len(u))]
     return sum(h)

def cube(n):
     v = {}
     G = Graph(2**n)
     vertices = Tuples([0,1],n)
     for i in range(2**n):
         for j in range(2**n):
             if dist(vertices[i],vertices[j]) == 1:
                    G.add_edge(i,j)
                    for x in range(n):
                        if vertices[i][x] == vertices[j][x]:
                            pass
                        else:
                            G.set_edge_label(i,j,x)
     for i in range(2**n):
        t = (tuple(vertices[i]))
        v[i] = ''.join([str(value) for value in t])
     G.relabel(v)
     return G

L = cube(6)
#L.show(edge_colors=L._color_by_label(d))

v_colours = {"white": [], 'grey': []}
for vertex in L.vertices():
    lst = [int(i) for i in vertex]
    if (sum(lst)%2) == 0:
        v_colours["white"].append(vertex)
    else:
        v_colours["grey"].append(vertex)

print("***************************************************")
print("vertices as either fermion (black) or boson (white)")
L.show(edge_colors=L._color_by_label(d),vertex_colors = v_colours)
```
